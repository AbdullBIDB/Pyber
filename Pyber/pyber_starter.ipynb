```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data

city = pd.read_csv(city_data_to_load)
city.head()
ride = pd.read_csv(ride_data_to_load)
ride.head()
# Combine the data into a single dataset
cityride = pd.merge(city, ride, on="city", how="outer")
cityride.head()
# Display the data table for preview
cityride_df = cityride[["city","date","fare","ride_id","driver_count","type"]]
cityride_dff = cityride[['type','city','fare','ride_id','driver_count']]
cityride_dff.head()

##This is for unite testing
#cityride_dff.to_csv("pyber_clean.csv",
#encoding="utf-8", index=False, header=True)
```

```python
#Rural City Type axis for the Bubble Plot
rural_df = cityride_dff.loc[cityride_dff["type"]== "Rural", :]
rural_df_g = rural_df.groupby(['city'])
rider_rc = rural_df_g.count()
driver_rc = rural_df_g.sum()
avg_fare_r = rural_df_g.mean()



#Suburban City Type axis for the Bubble Plot
suburban_df = cityride_dff.loc[cityride_dff["type"]== "Suburban", :]
suburban_df_g = suburban_df.groupby(['city'])
rider_sc = suburban_df_g.count()
driver_sc = suburban_df_g.sum()
avg_fare_s = suburban_df_g.mean()


#Urban City Type axis for the Bubble Plot
urban_df = cityride_dff.loc[cityride_dff["type"]== "Urban", :]
urban_df_g = urban_df.groupby(['city'])
rider_uc = urban_df_g.count()
driver_uc = urban_df_g.sum()
avg_fare_u = urban_df_g.mean()

```

## Bubble Plot of Ride Sharing Data

```python
#Scatter Plot for all three City Types and all their axis
plt.scatter(rider_uc['ride_id'], avg_fare_u['fare'], marker="o", facecolors="LightCoral", edgecolors="black",
            s=driver_uc['driver_count'], alpha=0.75, label="Urban", linewidths=1)


plt.scatter(rider_sc['ride_id'], avg_fare_s['fare'], marker="o", facecolors="LightSkyBlue", edgecolors="black",
            s=driver_sc['driver_count'], alpha=0.75, label="Suburban", linewidths=1)


plt.scatter(rider_rc['ride_id'], avg_fare_r['fare'], marker="o", facecolors="Gold", edgecolors="black",
            s=driver_rc['driver_count'], alpha=0.75, label="Rural", linewidths=1)

#Setting up the Scatter Plot (title, x & y labels)
plt.title("Pyber Ride Data 2016")
plt.xlabel("Total Number of Rides Per City")
plt.ylabel("Average Fare ($) Per City")
plt.tight_layout()
plt.grid()
lgnd = plt.legend(fontsize="small", loc='best', mode="expanded", numpoints=1, scatterpoints=1, labelspacing=0.5)
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]

plt.figure()
plt.show()
```

## Total Fares by City Type

```python
#Summing up all the Fares Data
total_far_avg = cityride_dff.groupby(['type'])
total_f = round(total_far_avg['fare'].sum(), 2)
far_pie = total_f.plot(kind="pie", y=total_f, explode=[0,0,0.1], shadow=True, startangle=150, autopct="%1.1f%%",\
                       colors=["Gold","LightSkyBlue","LightCoral"])
plt.axis("equal")
plt.title("% of Total Fares by City Type")
plt.savefig("Images/FareByCityType.png")
plt.show()
```

## Total Rides by City Type

```python
#Summing up all the Rides 
total_rider = cityride_dff.groupby(['type'])
total_r = total_rider["ride_id"].count()
rider_pie = total_r.plot(kind="pie", y=total_r, explode=[0,0,0.1], shadow=True, startangle=150, autopct="%1.1f%%", \
                         colors=["Gold","LightSkyBlue","LightCoral"])
plt.axis("equal")
plt.title("% of Total Rides by City Type")
plt.savefig("Images/RidesByCityType")
plt.show()
```

## Total Drivers by City Type

```python
#Summing up all the total Drivers by
clean_dub = cityride_dff[['type','city','driver_count']]
clean_new = clean_dub.drop_duplicates()
clean = clean_new.groupby(['type'])
final_driver = clean['driver_count'].sum()
driver_pie = final_driver.plot(kind="pie", y=final_driver, explode=[0,0,0.1], shadow=True, startangle=150, autopct="%1.1f%%", \
                               colors=["Gold","LightSkyBlue","LightCoral"])
plt.axis("equal")
plt.title("% of Total Drivers by City Type")
plt.savefig("Images/DriversByCityType")
plt.show()
```

Bubble Plot: Here is the finding from the Bubble Plot:

o	Because the Rural city type is far from the City/Urban area, it is low on the number of rides and drivers while it is high on the average fare

o	The Urban city type is high on number or rides plus number of Drivers while the average fare is low comparing to the Rural area and Suburban areas. 


Pie Charts:

-	The Urban city type area has the highest drivers and rides. Also, it has the highest average fare because of the volume of rides in the area.

-	The Rural city type area has the lowest drivers and rides. Even though the Rural area has the highest fares, however it did not have the highest total fare by the city type.
